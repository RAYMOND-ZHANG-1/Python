4.When too many users make Instacart orders at the same time, the app freezes. 
The senior technical officer at Instacart wants you to identify the busiest hours of the day. 
Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.”
Create a new column containing these labels called “busiest_period_of_day.”

```python
# Checking 'order_hour_of_day' value_counts
ord_pro['order_hour_of_day'].value_counts(dropna = False)
```

Answer:

First, I will create 3 groups from 24 hours of the day, which will met the question that asks for 'Most orders'
,'Average orders' and 'Fewest orders'

Most orders group (10, 11, 14, 15, 13, 12, 16, 9)

Average orders group (17, 8, 18, 19, 20, 7, 21, 22)

Fewest orders group (23, 6, 0, 1, 5, 2, 4, 3)

Second, using for-lop function to create conditions

```python
hour = []
for value in ord_pro['order_hour_of_day']:
    if value in  (10, 11, 14, 15, 13, 12, 16, 9):
        hour.append('Most orders')
    elif value in (17, 8, 18, 19, 20, 7, 21, 22):
        hour.append('Average orders')
    elif value in(23, 6, 0, 1, 5, 2, 4, 3):
        hour.append('Fewest orders')
    else:
        hour.append('No data')
```

```python
ord_pro['busiest_period_of_day'] = hour
```
